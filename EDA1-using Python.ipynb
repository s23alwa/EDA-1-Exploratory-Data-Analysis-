1.Data Retrieval

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set(color_codes=True)
```

```python
df1 = pd.read_csv('https://storage.googleapis.com/hj6418/dashboard_latest_data_2.csv')
df1.info()
df1.head()

```

```python
df1.describe()
```

```python
df1.nunique()
```

```python
df1.info(memory_usage='deep')
```

2.Data Cleaning

From the above data, we can observe that the column 'footnote'some missing values. Hence, we drop the column as it will not help in further analysis. Even though the column 'many_waves' has a single value, it will be ideal to have this column as one might add more data for the next year - 2023.

```python
df1 = df1.drop(columns="footnote") 
df1.info()
```

```python
df1.drop(['weight_type'], axis=1, inplace=True)
df1.info()
```

python visualization libraries can interpret the categorical data type to apply approrpiate statistical models or plot types. Categorical data uses less memory which can lead to performance improvements.

```python
df1["region_code"] = df1["region_code"].astype('category')
df1.info(memory_usage="deep")
df1.head()
```

```python
df1.columns = df1.columns.str.replace(' ', '_').str.lower()
df1.info()
```

```python
df1['many_waves'].fillna('falsy_values', inplace=True)
df1['many_waves'].unique
```

```python
#The gdp column presumably contains the country's GDP value across many years in ascending order and using linear regression, you can impute the missing values.
```

```python
df1['gdp_pc'].interpolate(method='linear', inplace=True)
```

```python
groups = df1.groupby(['month', 'year'])
print(groups.size())
```

```python
df1.dropna(axis = 0, how = 'any',subset = ['month','sample_total'],inplace = True)
df1.info()
```

```python
from numpy.ma.core import sort
df1.month = df1.month.astype(int)
df1.head()
```

3.Analysis








Finding the relation between variables

```python
plt.figure(figsize=(10,5))
c = df1.corr()
sns.heatmap(c, cmap="BrBG", annot=True)
c
```

From the above heat-map, we can see the highly correlated values are 1 close to 1 , negatively correlated values are -0.2 to -0.4,and there exits no correlation at 0

```python
ax = sns.barplot(x="region", y="gdp_pc",data=df1).set_title("countries that have effected economically during covid_19")
plt.xticks(rotation='vertical')
```

From above barplot,we can conclude that the sub-saharan africa region got effected worstly in relate with the GDP(econimical factors as concerned such as,inefficiency in production,gowth,unemployment,food scarcity etc.) as compared with the other regions
